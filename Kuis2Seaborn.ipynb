```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
gm = sns.load_dataset('tips')
```

```python
sns.set(style="darkgrid")
g = sns.FacetGrid(gm, col="smoker", hue='sex')
g.map(plt.hist, "total_bill", color='#56b4ff')
```
